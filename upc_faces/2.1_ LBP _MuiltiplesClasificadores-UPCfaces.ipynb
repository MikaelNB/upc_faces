```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import imshow

vec = []
labels = []
imagesPerLabel = 10

testVec = []
k = 0

for index in label_id:    
    for root, dirs, files in os.walk(new_input_path + label_id[index]):
        #print(root, dirs, files)
        for name in files:
            if len(pics[index]) > imagesPerLabel -1:
                print(index, len(pics[index]))
                break
            #print(name)
            if name.endswith((".jpg")):           
                img = Image.open(os.path.join(os.fsdecode(root), name), 'r').convert('L')
                #print(os.path.join(os.fsdecode(root), name))
                #gray = cv2.cvtColor(np.asarray(img), cv2.COLOR_RGB2GRAY)
                
                #gray = cv2.cvtColor(np.uint8(img), cv2.COLOR_RGB2GRAY)
                imshow(np.asarray(img), 'gray')                
                plt.show()
                lbp = feature.local_binary_pattern(np.asarray(img), 24,8, method="default")
                rgb = cv2.cvtColor(np.uint8(lbp), cv2.COLOR_GRAY2RGB)
                imshow(rgb, 'gray')
                plt.show()
                imagen = Image.fromarray(np.asarray(rgb))
                vec = img2vec.get_vec(imagen)
                if  testVec == [] and index == 0:
                    k += 1
                    if k == 5:
                        print('index: ', index)
                        testVec = vec
                print(vec.shape)                
                pics[index].append(vec)     
                #print(name)
                labels.append(int(index))
    pics[index] = np.array(pics[index])    
    
labels = np.array(labels)
```

```python
f1 = h5py.File('./UPCfaces_hdf5/data_UPCfaces_lbp.hdf5', 'r')
X = f1.get('dataset_1').value # `data` is now an ndarray.
f1.close()
#X = np.array(X)

f2 = h5py.File('./UPCfaces_hdf5/labels_UPCfaces_lbp.hdf5', 'r')
y = f2.get('dataset_1').value # `data` is now an ndarray.
f2.close()
#y = np.array(y)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state = 6)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
#print(X.shape, y.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
print(scores)
print(scores.mean())
```

```python
# search for an optimal value of K for KNN

# range of k we want to try
k_range = range(1, 10)
# empty list to store scores
k_scores = []

# 1. we will loop through reasonable values of k
for k in k_range:
    # 2. run KNeighborsClassifier with k neighbours
    knn = KNeighborsClassifier(n_neighbors=k)
    # 3. obtain cross_val_score for KNeighborsClassifier with k neighbours
    scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')
    # 4. append mean of scores for k neighbors to k_scores list
    k_scores.append(scores.mean())


print(k_scores)
```

```python
start_time = time.time()
clf = MLPClassifier(alpha=3.1)
clf.fit(X_train,y_train)

scores = cross_val_score(clf, X, y, cv=10, scoring='accuracy')
print(scores)
print(scores.mean())

print("%s ms" % ((time.time() - start_time)*1000))
```

```python
start_time = time.time()
clf = RandomForestClassifier(max_depth=250, n_estimators=200, max_features=5)
clf.fit(X_train,y_train)

scores = cross_val_score(clf, X, y, cv=10, scoring='accuracy')
print(scores)
print(scores.mean())

print("%s ms" % ((time.time() - start_time)*1000))
```

```python
start_time = time.time()
clf = KNeighborsClassifier(n_neighbors=1, p=1)
clf.fit(X_train,y_train)

scores = cross_val_score(clf, X, y, cv=10, scoring='accuracy')
print(scores)
print(scores.mean())

print("%s ms" % ((time.time() - start_time)*1000))
```

```python
start_time = time.time()
clf = DecisionTreeClassifier(max_depth=250, min_samples_split=2)
clf.fit(X_train,y_train)

scores = cross_val_score(clf, X, y, cv=10, scoring='accuracy')
print(scores)
print(scores.mean())

print("%s ms" % ((time.time() - start_time)*1000))
```
